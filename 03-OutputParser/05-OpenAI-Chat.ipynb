### í…œí”Œë¦¿ìœ¼ë¡œ ì–‘ì‹ ì •ì˜

```python
from langchain.memory import ConversationBufferMemory
from langchain.prompts import (
    ChatPromptTemplate,
    MessagesPlaceholder,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,
)
from langchain.chains import LLMChain


# 1) ì‹œìŠ¤í…œ ì„¤ì •: ì—­í• ë¶€ì—¬ ì •ì˜
system_template = SystemMessagePromptTemplate.from_template("ë„ˆëŠ” í•œê¸€ì„ {language}ë¡œ ë²ˆì—­í•´ ì£¼ëŠ” ë²ˆì—­ì „ë¬¸ê°€ì•¼.")
system_message = system_template.format(language="ì˜ì–´")

# 2) ChatPromptTemplate í…œí”Œë¦¿ ì •ì˜
prompt = ChatPromptTemplate.from_messages([
    system_message,                                              # ì—­í• ë¶€ì—¬
    MessagesPlaceholder(variable_name="chat_history"),           # ë©”ëª¨ë¦¬ ì €ì¥ì†Œ ì„¤ì •. ConversationBufferMemoryì˜ memory_key ì™€ ë™ì¼í•˜ê²Œ ì„¤ì •
    HumanMessagePromptTemplate.from_template("{human_input}"),   # ì‚¬ìš©ì ë©”ì‹œì§€ injection
])


# 3) ë©”ëª¨ë¦¬ ì •ì˜
# `return_messages=True`ë¥¼ ì‚¬ìš©í•´ì•¼ MessagesPlaceholderì— ì£¼ì…ë¨
# `chat_history`ê°€ MessagesPlaceholder ì´ë¦„ê³¼ ì¼ì¹˜í•´ì•¼í•¨
memory = ConversationBufferMemory(memory_key="chat_history", 
                                  ai_prefix="ë²ˆì—­ê°€",
                                  human_prefix="ì‚¬ìš©ì",
                                  return_messages=True)


# 4) LLM ëª¨ë¸ ì •ì˜
llm = ChatOpenAI(model_name='gpt-3.5-turbo')

# 5) LLMChain ì •ì˜
conversation = LLMChain(
    llm=llm,       # LLM
    prompt=prompt, # Prompt
    verbose=True,  # True ë¡œ ì„¤ì •ì‹œ ë¡œê·¸ ì¶œë ¥
    memory=memory  # ë©”ëª¨ë¦¬
)

# 6) ì‹¤í–‰
answer = conversation({"human_input": "ë‚˜ëŠ” í”„ë¡œê·¸ë˜ë°ì„ ì •ë§ ì¢‹ì•„í•´ìš”"})
print(answer['text'])
```

## ConversationSummaryBufferMemory: ìš”ì•½ ë©”ëª¨ë¦¬

`ConversationSummaryBufferMemory` ëŠ”ìµœê·¼ì˜ ëŒ€í™” ë‚´ìš©ì„ ë©”ëª¨ë¦¬ì— ì €ì¥í•©ë‹ˆë‹¤.

ë‹¤ë§Œ, ëŒ€í™”ê°€ ê¸¸ì–´ì§ˆ ê²½ìš° ì´ì „ ëŒ€í™”ë‚´ìš©ì„ ëª¨ë‘ í¬í•¨í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ë”°ë¼ì„œ, ì´ì „ ëŒ€í™” ë‚´ìš©ì„ ì‚­ì œí•˜ê²Œ ë˜ëŠ”ë°, `ConversationSummaryBufferMemory` ëŠ” ì™„ì „íˆ ì‚­ì œí•˜ê¸°ë³´ë‹¤ëŠ” ìš”ì•½ ì •ë³´ë¡œ ë§Œë“­ë‹ˆë‹¤.

ì´ì „ ëŒ€í™”ë‚´ìš©ì— ëŒ€í•œ ìš”ì•½ì´ ë°œë™í•˜ëŠ” ì‹œì ì€ **ëŒ€í™”ì˜ interaction íšŸìˆ˜ê°€ ì´ˆê³¼ëœ ê²½ìš° í˜¹ì€ í† í° ê¸¸ì´ê°€ ì´ˆê³¼ëœ ê²½ìš°**ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë™ì‘í•©ë‹ˆë‹¤.

```python
from langchain.memory import ConversationSummaryBufferMemory
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler


# 1) ì‹œìŠ¤í…œ ì„¤ì •: ì—­í• ë¶€ì—¬ ì •ì˜
system_template = SystemMessagePromptTemplate.from_template(
    "ë„ˆëŠ” íšŒì‚¬ì˜ ë§ˆì¼€íŒ… ë‹´ë‹¹ìì•¼. ë‹¤ìŒì˜ ìƒí’ˆ ì •ë³´ë¥¼ ë°”íƒ•ìœ¼ë¡œ ê³ ê° ì‘ëŒ€ë¥¼ í•´ì¤˜.\n{product}"
)
system_message = system_template.format(product="""
ìƒí’ˆëª…: í•˜ëŠ˜ì„ ë‚˜ëŠ” ìŠ¤ë§ˆíŠ¸í°
ì œí’ˆê°€ê²©: 300ë§Œì›
ì£¼ìš”ê¸°ëŠ¥:
- ìŠ¤ë§ˆíŠ¸í°ì„ ê³µì¤‘ì— ë„ìš¸ ìˆ˜ ìˆìŒ
- ë°©ìˆ˜ê¸°ëŠ¥
- ìµœì‹ ì‹ ì¸ê³µì§€ëŠ¥ ì†”ë£¨ì…˜ íƒ‘ì¬
- ì „í™”í†µí™”
- ìŒì•…
- ì¸í„°ë„· ë¸Œë¼ìš°ì§•, 5G ë„¤íŠ¸ì›Œí¬, ì™€ì´íŒŒì´, ë¸”ë£¨íˆ¬ìŠ¤
ì œì¡°ì‚¬: íŒŒì¸ì• í”Œ
""")

# 2) ChatPromptTemplate í…œí”Œë¦¿ ì •ì˜
prompt = ChatPromptTemplate.from_messages([
    system_message,                                              # ì—­í• ë¶€ì—¬
    MessagesPlaceholder(variable_name="chat_history"),           # ë©”ëª¨ë¦¬ ì €ì¥ì†Œ ì„¤ì •. ConversationBufferMemoryì˜ memory_key ì™€ ë™ì¼í•˜ê²Œ ì„¤ì •
    HumanMessagePromptTemplate.from_template("{human_input}"),   # ì‚¬ìš©ì ë©”ì‹œì§€ injection
])


# 3) ë©”ëª¨ë¦¬ ì •ì˜
memory = ConversationSummaryBufferMemory(llm=llm, 
                                         memory_key="chat_history", 
                                         max_token_limit=10, 
                                         return_messages=True
                                        )


# 4) LLM ëª¨ë¸ ì •ì˜
llm = ChatOpenAI(model_name='gpt-3.5-turbo', 
                 temperature=0.5,
                 streaming=True,
                 callbacks=[StreamingStdOutCallbackHandler()])

# 5) LLMChain ì •ì˜
conversation = LLMChain(
    llm=llm,       # LLM
    prompt=prompt, # Prompt
    verbose=True,  # True ë¡œ ì„¤ì •ì‹œ ë¡œê·¸ ì¶œë ¥
    memory=memory  # ë©”ëª¨ë¦¬
)

# 6) ì‹¤í–‰
answer = conversation({"human_input": "ì €ì—ê²Œ ì¶”ì²œí•´ ì¤„ë§Œí•œ ìƒí’ˆì´ ìˆë‚˜ìš”?"})
```
