## load_summarize_chain

ë¯¸ë¦¬ë³´ê¸°ë¥¼ ì œê³µí•˜ê¸° ìœ„í•´, ì–´ë–¤ íŒŒì´í”„ë¼ì¸ë„ ë‹¨ì¼ ê°ì²´ë¡œ ë˜í•‘ë  ìˆ˜ ìˆìŠµë‹ˆë‹¤: `load_summarize_chain`.

ë¸”ë¡œê·¸ í¬ìŠ¤íŠ¸ë¥¼ ìš”ì•½í•˜ê³  ì‹¶ë‹¤ê³  ê°€ì •í•´ ë´…ì‹œë‹¤. ëª‡ ì¤„ì˜ ì½”ë“œë¡œ ì´ë¥¼ ìƒì„±í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ë¨¼ì € í™˜ê²½ ë³€ìˆ˜ë¥¼ ì„¤ì •í•˜ê³  íŒ¨í‚¤ì§€ë¥¼ ì„¤ì¹˜í•˜ì„¸ìš”:


```python
# API KEYë¥¼ í™˜ê²½ë³€ìˆ˜ë¡œ ê´€ë¦¬í•˜ê¸° ìœ„í•œ ì„¤ì • íŒŒì¼
from dotenv import load_dotenv

# API KEY ì •ë³´ë¡œë“œ
load_dotenv()
```

`chain_type="stuff"`ë¥¼ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

`chain_type="map_reduce"` ë˜ëŠ” `chain_type="refine"`ë„ ì œê³µí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤(ë” ì½ì–´ë³´ê¸° [ì—¬ê¸°](/docs/modules/chains/document/refine)).


ì´ ì½”ë“œëŠ” ì›¹ì—ì„œ ë¬¸ì„œë¥¼ ë¡œë“œí•˜ê³ , ì´ë¥¼ ìš”ì•½í•˜ê¸° ìœ„í•´ `langchain` ë¼ì´ë¸ŒëŸ¬ë¦¬ì™€ OpenAIì˜ GPT ëª¨ë¸ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.

ë¨¼ì €, `WebBaseLoader`ë¥¼ ì‚¬ìš©í•˜ì—¬ ì§€ì •ëœ URLì—ì„œ ë¬¸ì„œë¥¼ ë¡œë“œí•©ë‹ˆë‹¤. ê·¸ ë‹¤ìŒ, `ChatOpenAI`ë¥¼ ì‚¬ìš©í•˜ì—¬ GPT-3.5 ëª¨ë¸ì„ ì´ˆê¸°í™”í•©ë‹ˆë‹¤.

`load_summarize_chain` í•¨ìˆ˜ë¥¼ í†µí•´ ìš”ì•½ ì‘ì—…ì„ ìœ„í•œ ì²´ì¸ì„ ë¡œë“œí•©ë‹ˆë‹¤.

ë§ˆì§€ë§‰ìœ¼ë¡œ, ë¡œë“œëœ ë¬¸ì„œì— ëŒ€í•´ ìš”ì•½ ì²´ì¸ì„ ì‹¤í–‰í•©ë‹ˆë‹¤. ì´ ê³¼ì •ì€ AIë¥¼ í™œìš©í•˜ì—¬ ì›¹ ë¬¸ì„œë¥¼ ìš”ì•½í•˜ëŠ” íš¨ìœ¨ì ì¸ ë°©ë²•ì„ ì œì‹œí•©ë‹ˆë‹¤.


```python
from langchain.chains.summarize import load_summarize_chain
from langchain_community.document_loaders import WebBaseLoader
from langchain_openai import ChatOpenAI
from langchain.callbacks.base import BaseCallbackHandler

# ì›¹ ê¸°ë°˜ ë¬¸ì„œ ë¡œë”ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤.
loader = WebBaseLoader("https://lilianweng.github.io/posts/2023-06-23-agent/")

# ë¬¸ì„œë¥¼ ë¡œë“œí•©ë‹ˆë‹¤.
docs = loader.load()


class StreamCallback(BaseCallbackHandler):
    def on_llm_new_token(self, token, **kwargs):
        print(f"{token}", end="", flush=True)


# OpenAIì˜ Chat ëª¨ë¸ì„ ì´ˆê¸°í™”í•©ë‹ˆë‹¤. ì—¬ê¸°ì„œëŠ” ì˜¨ë„ë¥¼ 0ìœ¼ë¡œ ì„¤ì •í•˜ê³  ëª¨ë¸ ì´ë¦„ì„ ì§€ì •í•©ë‹ˆë‹¤.
llm = ChatOpenAI(
    temperature=0,
    model_name="gpt-3.5-turbo-16k",
    streaming=True,
    callbacks=[StreamCallback()],
)
# ìš”ì•½ ì²´ì¸ì„ ë¡œë“œí•©ë‹ˆë‹¤. ì²´ì¸ íƒ€ì…ì„ 'stuff'ë¡œ ì§€ì •í•©ë‹ˆë‹¤.
chain = load_summarize_chain(llm, chain_type="stuff")

# ë¬¸ì„œì— ëŒ€í•´ ìš”ì•½ ì²´ì¸ì„ ì‹¤í–‰í•©ë‹ˆë‹¤.
answer = chain.invoke({"input_documents": docs})
print(answer["output_text"])
```

```python
from langchain.chains.combine_documents.stuff import StuffDocumentsChain
from langchain.chains.llm import LLMChain
from langchain.prompts import PromptTemplate
from langchain import hub

# ìš”ì•½ë¬¸ì„ ì‘ì„±í•˜ê¸° ìœ„í•œ í”„ë¡¬í”„íŠ¸ ì •ì˜ (ì§ì ‘ í”„ë¡¬í”„íŠ¸ë¥¼ ì‘ì„±í•˜ëŠ” ê²½ìš°)
# prompt_template = """Please summarize the sentence according to the following REQUEST.
# REQUEST:
# 1. Summarize the main points in bullet points in KOREAN.
# 2. Each summarized sentence must start with an emoji that fits the meaning of the each sentence.
# 3. Use various emojis to make the summary more interesting.
# 4. Translate the summary into Korean if it is written in English.
# 5. DO NOT translate any technical terms.
# 6. DO NOT include any unnecessary information.
# CONTEXT:
# {context}

# SUMMARY:"
# """
# prompt = PromptTemplate.from_template(prompt_template)

# ì›ê²© ì €ì¥ì†Œì—ì„œ í”„ë¡¬í”„íŠ¸ë¥¼ ê°€ì ¸ì˜¤ëŠ” ê²½ìš°
prompt = hub.pull("teddynote/summary-stuff-documents-korean")

# LLM ì²´ì¸ ì •ì˜
llm = ChatOpenAI(
    temperature=0,
    model_name="gpt-3.5-turbo-16k",
    streaming=True,
    callbacks=[StreamCallback()],
)

# LLMChain ì •ì˜
llm_chain = LLMChain(llm=llm, prompt=prompt)

# StuffDocumentsChain ì •ì˜
stuff_chain = StuffDocumentsChain(llm_chain=llm_chain, document_variable_name="context")

docs = loader.load()
response = stuff_chain.invoke({"input_documents": docs})
```

ì´ ë¬¸ì„œëŠ” `LLMChain`, `StuffDocumentsChain`, `ReduceDocumentsChain` í´ë˜ìŠ¤ë¥¼ ì‚¬ìš©í•˜ì—¬ ë¬¸ì„œ ì²˜ë¦¬ íŒŒì´í”„ë¼ì¸ì„ êµ¬ì„±í•˜ëŠ” ë°©ë²•ì„ ë³´ì—¬ì¤ë‹ˆë‹¤. `LLMChain`ì€ ì´ˆê¸° ì²˜ë¦¬ ë‹¨ê³„ë¡œ, íŠ¹ì • í”„ë¡¬í”„íŠ¸ë¥¼ ì‚¬ìš©í•˜ì—¬ ì–¸ì–´ ëª¨ë¸(`llm`)ì„ ì‹¤í–‰í•©ë‹ˆë‹¤. `StuffDocumentsChain`ì€ ì—¬ëŸ¬ ë¬¸ì„œë¥¼ í•˜ë‚˜ì˜ ë¬¸ìì—´ë¡œ ê²°í•©í•˜ì—¬ `LLMChain`ì— ì „ë‹¬í•˜ëŠ” ì—­í• ì„ í•©ë‹ˆë‹¤. ë§ˆì§€ë§‰ìœ¼ë¡œ, `ReduceDocumentsChain`ì€ ë¬¸ì„œë“¤ì„ ê²°í•©í•˜ê³ , ì§€ì •ëœ í† í° ìˆ˜(`token_max`)ë¥¼ ì´ˆê³¼í•˜ì§€ ì•Šë„ë¡ ë°˜ë³µì ìœ¼ë¡œ ì¶•ì†Œí•˜ëŠ” ê³¼ì •ì„ ë‹´ë‹¹í•©ë‹ˆë‹¤. ì´ ê³¼ì •ì—ì„œ, ë¬¸ì„œë“¤ì´ `StuffDocumentsChain`ì˜ ì»¨í…ìŠ¤íŠ¸ë¥¼ ì´ˆê³¼í•  ê²½ìš°, ë™ì¼í•œ ì²´ì¸(`collapse_documents_chain`)ì„ ì‚¬ìš©í•˜ì—¬ ì²˜ë¦¬í•©ë‹ˆë‹¤.


```python
# ì—°ì‡„ ì‹¤í–‰
reduce_chain = LLMChain(llm=llm, prompt=reduce_prompt)

# ë¬¸ì„œ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°›ì•„ í•˜ë‚˜ì˜ ë¬¸ìì—´ë¡œ ê²°í•©í•œ í›„ LLMChainì— ì „ë‹¬
combine_documents_chain = StuffDocumentsChain(
    llm_chain=reduce_chain, document_variable_name="doc_summaries"
)

# ë§¤í•‘ëœ ë¬¸ì„œë“¤ì„ ê²°í•©í•˜ê³  ë°˜ë³µì ìœ¼ë¡œ ì¶•ì†Œ
reduce_documents_chain = ReduceDocumentsChain(
    # ìµœì¢…ì ìœ¼ë¡œ í˜¸ì¶œë˜ëŠ” ì²´ì¸ì…ë‹ˆë‹¤.
    combine_documents_chain=combine_documents_chain,
    # `StuffDocumentsChain`ì˜ ì»¨í…ìŠ¤íŠ¸ë¥¼ ì´ˆê³¼í•˜ëŠ” ë¬¸ì„œë“¤ì„ ì²˜ë¦¬
    collapse_documents_chain=combine_documents_chain,
    # ë¬¸ì„œë“¤ì„ ê·¸ë£¹í™”í•  ìµœëŒ€ í† í° ìˆ˜.
    token_max=4096,
)
```

ìš°ë¦¬ì˜ mapê³¼ reduce ì²´ì¸ì„ í•˜ë‚˜ë¡œ ê²°í•©í•´ ë´…ì‹œë‹¤.


ì´ ì½”ë“œëŠ” ë¬¸ì„œë“¤ì„ ë§¤í•‘í•˜ê³  ë¦¬ë“€ìŠ¤í•˜ëŠ” ê³¼ì •ì„ í†µí•´ ê²°í•©í•˜ëŠ” `MapReduceDocumentsChain` ê°ì²´ë¥¼ ìƒì„±í•˜ê³ , ë¬¸ì ê¸°ë°˜ìœ¼ë¡œ í…ìŠ¤íŠ¸ë¥¼ ë¶„í• í•˜ëŠ” `CharacterTextSplitter` ê°ì²´ë¥¼ ì‚¬ìš©í•˜ì—¬ ë¬¸ì„œë“¤ì„ ë¶„í• í•©ë‹ˆë‹¤. `MapReduceDocumentsChain`ì€ ë§¤í•‘ ì²´ì¸(`llm_chain`), ë¦¬ë“€ìŠ¤ ì²´ì¸(`reduce_documents_chain`), ë¬¸ì„œë¥¼ ì €ì¥í•  ë³€ìˆ˜ ì´ë¦„(`document_variable_name`), ê·¸ë¦¬ê³  ë§¤í•‘ ë‹¨ê³„ì˜ ì¤‘ê°„ ê²°ê³¼ë¥¼ ë°˜í™˜í• ì§€ ì—¬ë¶€(`return_intermediate_steps`)ë¥¼ ì„¤ì •í•˜ì—¬ ì´ˆê¸°í™”ë©ë‹ˆë‹¤. `CharacterTextSplitter`ëŠ” `from_tiktoken_encoder` ë©”ì†Œë“œë¥¼ í†µí•´ ì´ˆê¸°í™”ë˜ë©°, ì´ëŠ” ë¶„í• í•  ì²­í¬ì˜ í¬ê¸°(`chunk_size`)ì™€ ì²­í¬ ê°„ ê²¹ì¹¨(`chunk_overlap`)ì„ ì„¤ì •í•©ë‹ˆë‹¤.


```python
# ë¬¸ì„œë“¤ì„ ë§¤í•‘í•˜ì—¬ ì²´ì¸ì„ ê±°ì¹œ í›„ ê²°ê³¼ë¥¼ ê²°í•©í•˜ëŠ” ê³¼ì •
map_reduce_chain = MapReduceDocumentsChain(
    # ë§¤í•‘ ì²´ì¸
    llm_chain=map_chain,
    # ë¦¬ë“€ìŠ¤ ì²´ì¸
    reduce_documents_chain=reduce_documents_chain,
    # llm_chainì—ì„œ ë¬¸ì„œë“¤ì„ ë„£ì„ ë³€ìˆ˜ ì´ë¦„
    document_variable_name="documents",
    # ë§¤í•‘ ë‹¨ê³„ì˜ ê²°ê³¼ë¥¼ ì¶œë ¥ì— í¬í•¨ì‹œí‚´
    return_intermediate_steps=False,
)

# ë¬¸ìë¥¼ ê¸°ì¤€ìœ¼ë¡œ í…ìŠ¤íŠ¸ë¥¼ ë¶„í• í•˜ëŠ” ê°ì²´ ìƒì„±
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=50,
    separators=["\n\n", "\n", "(?<=\. )", " ", ""],
    length_function=len,
)

# ë¬¸ì„œë“¤ì„ ë¶„í• 
split_docs = text_splitter.split_documents(docs)
```

`map_reduce_chain.run(split_docs)`ëŠ” `split_docs`ë¥¼ ì¸ìë¡œ ë°›ì•„ `map_reduce_chain`ì˜ `run` ë©”ì„œë“œë¥¼ ì‹¤í–‰í•˜ê³ , ê·¸ ê²°ê³¼ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤. ì´ëŠ” MapReduce íŒ¨í„´ì„ í™œìš©í•˜ì—¬ ë°ì´í„°ë¥¼ ì²˜ë¦¬í•˜ëŠ” ê³¼ì •ì„ ê°„ëµí•˜ê²Œ ë³´ì—¬ì¤ë‹ˆë‹¤. ì—¬ê¸°ì„œ `split_docs`ëŠ” ì²˜ë¦¬í•  ë°ì´í„°ë¥¼ ë‚˜íƒ€ë‚´ë©°, `map_reduce_chain`ì€ í•´ë‹¹ ë°ì´í„°ì— ì ìš©í•  MapReduce ì—°ì‚°ì˜ ì²´ì¸ì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.


```python
# split_docsë¥¼ map_reduce_chainì˜ run ë©”ì„œë“œì— ì „ë‹¬í•˜ì—¬ ì‹¤í–‰í•œ ê²°ê³¼ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
summary_result = map_reduce_chain.invoke({"input_documents": split_docs})
```

## ë°©ë²•3. Refine

![](./images/summarization_use_case_3.png)

`RefineDocumentsChain`ì€ map-reduceì™€ ìœ ì‚¬í•©ë‹ˆë‹¤:

> Refine documents chainì€ ì…ë ¥ ë¬¸ì„œë¥¼ ìˆœíšŒí•˜ë©° ë°˜ë³µì ìœ¼ë¡œ ë‹µë³€ì„ ì—…ë°ì´íŠ¸í•˜ì—¬ ì‘ë‹µì„ êµ¬ì„±í•©ë‹ˆë‹¤. ê° ë¬¸ì„œì— ëŒ€í•´, ëª¨ë“  ë¹„ë¬¸ì„œ ì…ë ¥, í˜„ì¬ ë¬¸ì„œ, ê·¸ë¦¬ê³  ìµœì‹  ì¤‘ê°„ ë‹µë³€ì„ LLM chainì— ì „ë‹¬í•˜ì—¬ ìƒˆë¡œìš´ ë‹µë³€ì„ ì–»ìŠµë‹ˆë‹¤.

ì´ëŠ” `chain_type="refine"`ì´ ì§€ì •ë˜ì–´ ìˆìœ¼ë©´ ì‰½ê²Œ ì‹¤í–‰í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.


ì´ í•¨ìˆ˜ëŠ” `load_summarize_chain`ì„ ì‚¬ìš©í•˜ì—¬ íŠ¹ì • ìœ í˜•ì˜ ìš”ì•½ ì²´ì¸ì„ ë¡œë“œí•˜ê³ , ì´ë¥¼ `run` ë©”ì†Œë“œë¥¼ í†µí•´ ì‹¤í–‰í•©ë‹ˆë‹¤. ì—¬ê¸°ì„œ `llm`ì€ ì–¸ì–´ ëª¨ë¸ì„ ë‚˜íƒ€ë‚´ë©°, `chain_type="refine"`ì€ ìš”ì•½ ê³¼ì •ì—ì„œ ì„¸ë¶€ ì¡°ì •ì„ ìœ„í•œ ì²´ì¸ ìœ í˜•ì„ ì§€ì •í•©ë‹ˆë‹¤. `split_docs`ëŠ” ì²˜ë¦¬í•  ë¬¸ì„œë“¤ì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.


```python
# llmì„ ì‚¬ìš©í•˜ì—¬ 'refine' ìœ í˜•ì˜ ìš”ì•½ ì²´ì¸ì„ ë¡œë“œí•©ë‹ˆë‹¤.
chain = load_summarize_chain(llm, chain_type="refine")
# split_docsë¥¼ ì²˜ë¦¬í•˜ê¸° ìœ„í•´ ì²´ì¸ì„ ì‹¤í–‰í•©ë‹ˆë‹¤.
chain.run(split_docs)
```

í”„ë¡¬í”„íŠ¸ë¥¼ ì œê³µí•˜ê³  ì¤‘ê°„ ë‹¨ê³„ë¥¼ ë°˜í™˜í•˜ëŠ” ê²ƒë„ ê°€ëŠ¥í•©ë‹ˆë‹¤.


`Refine` ë°©ë²•ìœ¼ë¡œ í…ìŠ¤íŠ¸ ìš”ì•½ ì‘ì—…ì„ ìœ„í•œ í”„ë¡œì„¸ìŠ¤ë¥¼ ì„¤ì •í•©ë‹ˆë‹¤.

`PromptTemplate.from_template` ë©”ì†Œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ìš”ì•½ ë° ìš”ì•½ ë‹¤ë“¬ê¸° ì‘ì—…ì— ì‚¬ìš©ë  í…œí”Œë¦¿ì„ ìƒì„±í•©ë‹ˆë‹¤.

`load_summarize_chain` í•¨ìˆ˜ëŠ” ìš”ì•½ ìƒì„± ë° ë‹¤ë“¬ê¸° ê³¼ì •ì„ ê´€ë¦¬í•˜ëŠ” ì²´ì¸ì„ ë¡œë“œí•©ë‹ˆë‹¤. ì´ ì²´ì¸ì€ ì´ˆê¸° ìš”ì•½ ìƒì„±(`prompt`)ê³¼ ê¸°ì¡´ ìš”ì•½ì˜ ê°œì„ (`refine_prompt`) ë‹¨ê³„ë¥¼ í¬í•¨í•©ë‹ˆë‹¤.

ë§ˆì§€ë§‰ìœ¼ë¡œ, `chain` í•¨ìˆ˜ëŠ” ì£¼ì–´ì§„ ë¬¸ì„œ(`input_documents`)ì— ëŒ€í•œ ìµœì¢… ìš”ì•½ ê²°ê³¼ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.


```python
prompt_template = """Write a concise summary of the following:
{text}
CONCISE SUMMARY:"""
prompt = PromptTemplate.from_template(prompt_template)

refine_template = (
    "Your job is to produce a final summary\n"
    "We have provided an existing summary up to a certain point: {existing_answer}\n"
    "We have the opportunity to refine the existing summary"
    "(only if needed) with some more context below.\n"
    "------------\n"
    "{text}\n"
    "------------\n"
    "Given the new context, refine the original summary in Korean"
    "If the context isn't useful, return the original summary."
)
refine_prompt = PromptTemplate.from_template(refine_template)
chain = load_summarize_chain(
    llm=llm,
    chain_type="refine",
    question_prompt=prompt,
    refine_prompt=refine_prompt,
    return_intermediate_steps=True,
    input_key="input_documents",
    output_key="output_text",
)
result = chain.invoke({"input_documents": split_docs}, return_only_outputs=True)
```

`print(result["output_text"])`ëŠ” ê²°ê³¼ ë”•ì…”ë„ˆë¦¬ `result`ì—ì„œ `'output_text'` í‚¤ì— í•´ë‹¹í•˜ëŠ” ê°’ì„ ì¶œë ¥í•©ë‹ˆë‹¤. ì´ êµ¬ë¬¸ì€ ë”•ì…”ë„ˆë¦¬ ë‚´ íŠ¹ì • í‚¤ì˜ ê°’ì„ ê²€ìƒ‰í•˜ê³ , ê·¸ ê°’ì„ ì½˜ì†”ì— ì¶œë ¥í•˜ëŠ” ê¸°ë³¸ì ì¸ Python ì½”ë“œ ì˜ˆì‹œì…ë‹ˆë‹¤.


```python
print(
    result["output_text"]
)  # ê²°ê³¼ ë”•ì…”ë„ˆë¦¬ì—ì„œ 'output_text' í‚¤ì— í•´ë‹¹í•˜ëŠ” ê°’ì„ ì¶œë ¥í•©ë‹ˆë‹¤.
```

ì´ í•¨ìˆ˜ëŠ” `result` ë”•ì…”ë„ˆë¦¬ì˜ `'intermediate_steps'` í‚¤ì— ì €ì¥ëœ ë¦¬ìŠ¤íŠ¸ì—ì„œ ì²˜ìŒ ì„¸ ìš”ì†Œë¥¼ ì„ íƒí•˜ê³ , ì´ë“¤ ì‚¬ì´ì— ë‘ ì¤„ë°”ê¿ˆ(`\n\n`)ì„ ì‚½ì…í•˜ì—¬ ì—°ê²°í•œ ë¬¸ìì—´ì„ ì¶œë ¥í•©ë‹ˆë‹¤. ì´ëŠ” ì¤‘ê°„ ê³„ì‚° ë‹¨ê³„ë‚˜ ê²°ê³¼ë¥¼ ì‹œê°ì ìœ¼ë¡œ êµ¬ë¶„í•˜ì—¬ í‘œì‹œí•  ë•Œ ìœ ìš©í•©ë‹ˆë‹¤.


```python
print("\n\n".join(result["intermediate_steps"][:3]))
```

## í•œ ë²ˆì˜ ì²´ì¸ìœ¼ë¡œ ë¶„í• í•˜ê³  ìš”ì•½í•˜ê¸°

í¸ì˜ë¥¼ ìœ„í•´, ìš°ë¦¬ëŠ” ê¸´ ë¬¸ì„œì˜ í…ìŠ¤íŠ¸ ë¶„í• ê³¼ ìš”ì•½ì„ ë‹¨ì¼ `AnalyzeDocumentsChain`ìœ¼ë¡œ ë¬¶ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.


`AnalyzeDocumentChain` í´ë˜ìŠ¤ëŠ” ë¬¸ì„œ ë¶„ì„ ë° ìš”ì•½ ì‘ì—…ì„ ìœ„í•œ ì²´ì¸ì„ ìƒì„±í•©ë‹ˆë‹¤. ì´ ì˜ˆì œì—ì„œëŠ” `AnalyzeDocumentChain` ì¸ìŠ¤í„´ìŠ¤ë¥¼ ìƒì„±í•˜ê³ , `combine_docs_chain`ê³¼ `text_splitter`ë¥¼ ì¸ìë¡œ ì „ë‹¬í•˜ì—¬ ì´ˆê¸°í™”í•©ë‹ˆë‹¤. ì´í›„, ì²« ë²ˆì§¸ ë¬¸ì„œ(`docs[0]`)ì˜ `page_content`ë¥¼ ì‚¬ìš©í•˜ì—¬ ë¬¸ì„œ ìš”ì•½ í”„ë¡œì„¸ìŠ¤ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤. ì´ ê³¼ì •ì€ ë¬¸ì„œì˜ ë‚´ìš©ì„ ë¶„ì„í•˜ê³  ìš”ì•½í•˜ëŠ” ë° ì‚¬ìš©ë©ë‹ˆë‹¤.


```python
from langchain.chains import AnalyzeDocumentChain

# AnalyzeDocumentChain ì¸ìŠ¤í„´ìŠ¤ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. ì´ë•Œ, combine_docs_chainê³¼ text_splitterë¥¼ ì¸ìë¡œ ì „ë‹¬í•©ë‹ˆë‹¤.
summarize_document_chain = AnalyzeDocumentChain(
    combine_docs_chain=chain, text_splitter=text_splitter
)
# ì²« ë²ˆì§¸ ë¬¸ì„œì˜ í˜ì´ì§€ ë‚´ìš©ì„ ì‚¬ìš©í•˜ì—¬ ë¬¸ì„œ ìš”ì•½ í”„ë¡œì„¸ìŠ¤ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤.
summarized_result = summarize_document_chain.invoke(
    {"input_document": docs[0].page_content}
)
```
