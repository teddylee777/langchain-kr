## ConversationChain


`ChatOpenAI` í´ë˜ìŠ¤ë¥¼ ì‚¬ìš©í•˜ì—¬ ëŒ€í™”í˜• AI ëª¨ë¸ì˜ ì¸ìŠ¤í„´ìŠ¤ë¥¼ ìƒì„±í•˜ê³ , ì´ë¥¼ í†µí•´ ì‚¬ìš©ìì™€ì˜ ëŒ€í™”ë¥¼ ì²˜ë¦¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
from langchain.chains import ConversationChain
from langchain_openai import ChatOpenAI
from langchain.memory import ConversationBufferMemory
```

`ConversationChain` í´ë˜ìŠ¤ì˜ ì¸ìŠ¤í„´ìŠ¤ë¥¼ ìƒì„±í•˜ì—¬ ëŒ€í™” ê´€ë¦¬ë¥¼ ìˆ˜í–‰í•©ë‹ˆë‹¤. ì´ë•Œ, `llm`ì€ ì €ìˆ˜ì¤€ ì–¸ì–´ ëª¨ë¸ì„, `verbose`ëŠ” ìƒì„¸í•œ ë¡œê¹… ì—¬ë¶€ë¥¼ (ì—¬ê¸°ì„œëŠ” ë¹„í™œì„±í™”), ê·¸ë¦¬ê³  `memory`ëŠ” ëŒ€í™” ë‚´ìš©ì„ ì €ì¥í•  ë©”ëª¨ë¦¬ ë²„í¼ë¥¼ ê°ê° ì§€ì •í•©ë‹ˆë‹¤.


```python
# ChatOpenAI ê°ì²´ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
llm = ChatOpenAI(model="gpt-4-turbo-preview")

# ConversationChain ì¸ìŠ¤í„´ìŠ¤ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
# llm: ëª¨ë¸ì„ ì§€ì •í•©ë‹ˆë‹¤.
# verbose: ìƒì„¸í•œ ë¡œê¹…ì„ ë¹„í™œì„±í™”í•©ë‹ˆë‹¤.
# memory: ëŒ€í™” ë‚´ìš©ì„ ì €ì¥í•˜ëŠ” ë©”ëª¨ë¦¬ ë²„í¼ë¥¼ ì§€ì •í•©ë‹ˆë‹¤.
conversation = ConversationChain(
    llm=llm,
    verbose=False,
    memory=ConversationBufferMemory(memory_key="history"),
)
```

## streaming: ì‹¤ì‹œê°„ ì¶œë ¥


ìŠ¤íŠ¸ë¦¬ë° ë°©ì‹ìœ¼ë¡œ ëŒ€í™”í˜• AI ëª¨ë¸ì„ êµ¬í˜„í•©ë‹ˆë‹¤.

`MyCustomHandler` í´ë˜ìŠ¤ëŠ” `BaseCallbackHandler`ë¥¼ ìƒì†ë°›ì•„, ìƒˆë¡œìš´ í† í°ì´ ìƒì„±ë  ë•Œë§ˆë‹¤ ì´ë¥¼ ì¶œë ¥í•˜ëŠ” ê¸°ëŠ¥ì„ ì •ì˜í•©ë‹ˆë‹¤.

`ChatOpenAI` ì˜ ìŠ¤íŠ¸ë¦¬ë° ëª¨ë“œë¥¼ í™œì„±í™”í•˜ê³ , ì‚¬ìš©ì ì •ì˜ í•¸ë“¤ëŸ¬ì¸ `MyCustomHandler` ì¸ìŠ¤í„´ìŠ¤ë¥¼ ì½œë°±ìœ¼ë¡œ ì„¤ì •í•©ë‹ˆë‹¤.

- [ì°¸ê³  ë¬¸ì„œ](https://python.langchain.com/docs/modules/callbacks/custom_callbacks)


```python
from langchain.callbacks.base import BaseCallbackHandler
from langchain_openai import ChatOpenAI


class MyCustomHandler(BaseCallbackHandler):
    def on_llm_new_token(self, token: str, **kwargs) -> None:
        print(f"{token}", end="", flush=True)


# ìŠ¤íŠ¸ë¦¬ë°ì„ í™œì„±í™”í•˜ê¸° ìœ„í•´, ChatModel ìƒì„±ìì— `streaming=True`ë¥¼ ì „ë‹¬í•©ë‹ˆë‹¤.
# ì¶”ê°€ì ìœ¼ë¡œ, ì‚¬ìš©ì ì •ì˜ í•¸ë“¤ëŸ¬ ë¦¬ìŠ¤íŠ¸ë¥¼ ì „ë‹¬í•©ë‹ˆë‹¤.
stream_llm = ChatOpenAI(
    model="gpt-4-turbo-preview", streaming=True, callbacks=[MyCustomHandler()]
)

conversation = ConversationChain(
    llm=stream_llm,
    verbose=False,
    memory=ConversationBufferMemory(),
)
```

`conversation.predict` ë©”ì†Œë“œëŠ” ì£¼ì–´ì§„ ì…ë ¥ì— ëŒ€í•œ ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤. ì´ ì˜ˆì‹œì—ì„œëŠ” 'ì–‘ìì—­í•™ì— ëŒ€í•´ ì„¤ëª…í•´ì¤˜'ë¼ëŠ” ë¬¸ìì—´ì„ ì…ë ¥ìœ¼ë¡œ ì‚¬ìš©í•˜ë©°, í•´ë‹¹ ì…ë ¥ì— ëŒ€í•œ ì˜ˆì¸¡ ê²°ê³¼ë¥¼ `output` ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤. ì´ ê³¼ì •ì€ AI ê¸°ë°˜ ëŒ€í™” ì‹œìŠ¤í…œì—ì„œ íŠ¹ì • ì£¼ì œì— ëŒ€í•œ ì„¤ëª…ì„ ìš”ì²­í•˜ëŠ” ë°©ë²•ì„ ë³´ì—¬ì¤ë‹ˆë‹¤.


```python
output = conversation.predict(input="ì–‘ìì—­í•™ì— ëŒ€í•´ ì„¤ëª…í•´ì¤˜")
```

```python
from langchain.prompts import PromptTemplate

template = """
ë‹¹ì‹ ì€ 10ë…„ì°¨ ì—‘ì…€ ì „ë¬¸ê°€ ì…ë‹ˆë‹¤. ì•„ë˜ ëŒ€í™”ë‚´ìš©ì„ ë³´ê³  ì§ˆë¬¸ì— ëŒ€í•œ ì ì ˆí•œ ë‹µë³€ì„ í•´ì£¼ì„¸ìš”

#ëŒ€í™”ë‚´ìš©
{chat_history}
----
ì‚¬ìš©ì: {question}
ì—‘ì…€ì „ë¬¸ê°€:"""

prompt = PromptTemplate.from_template(template)
```

## Human Prefix & AI Prefix ë³€ê²½

```python
template = """
ë‹¹ì‹ ì€ 10ë…„ì°¨ ì—‘ì…€ ì „ë¬¸ê°€ ì…ë‹ˆë‹¤. ì•„ë˜ ëŒ€í™”ë‚´ìš©ì„ ë³´ê³  ì§ˆë¬¸ì— ëŒ€í•œ ì ì ˆí•œ ë‹µë³€ì„ í•´ì£¼ì„¸ìš”

#ëŒ€í™”ë‚´ìš©
{context}
----
ğŸŒ±ê³ ê°: {question}
ğŸ¤–ì „ë¬¸ê°€:"""

prompt = PromptTemplate.from_template(template)
```
